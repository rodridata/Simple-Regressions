## Import the relevant libraries

```python
#Import libraries for graphs and stats.
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

## Load the data

```python
#Read the file
data = pd.read_csv("Real_estate_price_size.csv")
```

```python
data.describe()
```

### Declare the dependent and the independent variables

```python
#Declare variables
x1 = data['size']
y = data['price']
```

### Explore the data

```python
#Create plot graph to visualize the data
plt.scatter(x1,y)

plt.title('Description',size=30)
plt.xlabel('Size',size=20)
plt.ylabel('Prize',size=20)
```

```python
#REGRESSION ITSELF
```

```python
#Visualize data 

#Creates a column to estimate the interception.
x=sm.add_constant(x1)

#Generates the regression according to the results. This creates the line adapted to the data. 
results = sm.OLS(y,x).fit()

results.summary()
```

```python
print (y)
```

### Plot the regression line on the initial scatter

```python
plt.scatter(x1,y, color = 'darkblue')
yhat = 223.1787 * x1 + 1.019e+05

fig = plt.plot(x1, yhat, lw=4, c='red', label='regression line')

# Cambiar el color de fondo
plt.gca().set_facecolor('lightgray')

plt.xlabel('Size', fontsize = 20)
plt.ylabel('Prize', fontsize = 20)
plt.title('Simple Regression',fontsize=25)
plt.show()
```
